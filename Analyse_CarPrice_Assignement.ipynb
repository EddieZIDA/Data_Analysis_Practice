```python
# Importation et affichage des données 
import pandas as pd 
df=pd.read_csv('CarPrice_Assignement.csv')
df
```

```python
# Colonnes du dataframe
df.columns
```

```python
# Tri des voitures selon le prix, du plus cher au moins cher: les 10 premiers
df.sort_values(by='price', ascending=False).head(10)
```

```python
# les véhicules à essence avec plus de 150 chevaux 
df[(df['fueltype'] == 'gas') & (df['horsepower'] > 150)]
```

```python
# les voitures 4 portes de type berline
df[(df['doornumber'] == 'four') & (df['carbody'] == 'sedan')]
```

```python
# Renommons certaines colonnes
df.rename(columns={'car_ID': 'id','CarName': 'nom_voiture','cylindernumber': 'nb_cylindres','horsepower': 'puissance_cv',}, inplace=True)
df
```

```python
# Sélection des colonnes de type numérique
df.select_dtypes(include='number').columns
```

```python
# Détection de corrélation avec le prix
corr = df.corr(numeric_only=True)['price'].sort_values(ascending=False)
print(corr)
```

```python
# Moyenne des prix par carbody
df.groupby('carbody')['price'].mean().sort_values(ascending=False)
```

```python
# Moyenne, minimum, maximum du prix de carburant
df.groupby('fueltype')['price'].agg(['mean', 'min', 'max'])
```

```python
# Statistiques générales par drivewheel
df.groupby('drivewheel')[['puissance_cv', 'enginesize', 'price']].describe()
```

```python
# Compter le nombre de modèles par marque
df['marque'] = df['nom_voiture'].str.split().str[0]
df.groupby('marque').size().sort_values(ascending=False)
```

```python
# Création de la variable performance
df['performance'] = df['puissance_cv'] * df['enginesize']
df['performance']
```

```python
# Changeons le type de la colonne doornumber en entier
df['doornumber'] = df['doornumber'].map({'two': 2, 'four': 4})
df['doornumber'] 
```

```python
# Création de catégorie selon le  prix ( bas , moyen , elevé)
def classement_prix(cp):
    if cp < 15000:
        return 'bas'
    elif cp < 25000:
        return 'moyen'
    else:
        return 'élevé'

df['categorie_prix'] = df['price'].apply(classement_prix)
df['categorie_prix']
```

```python
# Création d'une colonne label (sport ou standard)
df['label'] = df.apply(lambda row: 'sport' if row['puissance_cv'] > 180 and row['carbody'] == 'coupe' else 'standard',axis=1)
df['label']
```

```python
# Analyse de corrélation
import seaborn as sns
import matplotlib.pyplot as plt
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Corrélation entre variables")
plt.show()
```
