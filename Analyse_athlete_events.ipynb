```python
# Ã‰volution du nombre de femmes participantes par annÃ©e
df_femmes = df[df["Sex"] == "F"]
df_femmes.groupby("Year")["ID"].nunique().plot(kind="line", figsize=(10,5), marker="o")
plt.title("Ã‰volution du nombre de femmes participantes")
plt.xlabel("AnnÃ©e")
plt.ylabel("Nombre d'athlÃ¨tes uniques")
plt.grid(True)
plt.show()

```
