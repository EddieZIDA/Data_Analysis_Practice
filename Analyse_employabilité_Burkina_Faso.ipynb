```python
# Identification des indicateurs
df['indicateur'].dropna().unique()
```

```python
# Il y'a donc au total 10 indicateurs, filtrons le dataframe en plusieurs dataframes selon les indicateurs

# Effectif des demandeurs d’emplois enregistrés à l'ANPE par région selon le sexe
df_demande_region_sexe = df[df['indicateur'] == "Effectif des demandeurs d’emplois enregistrés à l'ANPE par région selon le sexe"]

# Effectif des demandeurs d’emplois enregistrés à l'ANPE selon la tranche d’âge (en années révolues)
df_demande_age = df[df['indicateur'] == "Effectif des demandeurs d’emplois enregistrés à l'ANPE selon la tranche d’âge (en années révolues)"]

# Effectif des demandeurs d’emploi  enregistrés à l'ANPE selon le diplôme/titre de qualification et le sexe
df_demande_diplome = df[df['indicateur'] == "Effectif des demandeurs d’emploi  enregistrés à l'ANPE selon le diplôme/titre de qualification et le sexe"]

# Nombre de postes d'emplois publiés sur le site de l’ONEF selon la région d'affectation
df_postes_region = df[df['indicateur'] == "Nombre de postes d'emplois publiés sur le site de l’ONEF selon la région d'affectation"]

# Nombre de postes d’emplois publiés sur le site de l’ONEF selon le diplôme/titre de qualification
df_postes_diplome = df[df['indicateur'] == "Nombre de postes d’emplois publiés sur le site de l’ONEF selon le diplôme/titre de qualification"]

# Montant des financements sollicités (auprès du FASI, FAPE, FAIJ) par région selon le sexe (en millions de F CFA)
df_financement_solicite = df[df['indicateur'] == "Montant des financements sollicités (auprès du FASI, FAPE, FAIJ) par région selon le sexe (en millions de F CFA)"]

# Montant des financements accordés (par le FAPE, le FASI, le FAIJ et le PPEJ) par région selon le sexe (en millions de FCFA)
df_financement_accorde = df[df['indicateur'] == "Montant des financements accordés (par le FAPE, le FASI, le FAIJ et le PPEJ) par région selon le sexe (en millions de FCFA)"]

# Nombre de projets financés  (par le FAPE, le FASI, le FAIJ, le PPEJ et JOB BOOSTER) par région
df_projets_finances = df[df['indicateur'] == "Nombre de projets financés  (par le FAPE, le FASI, le FAIJ, le PPEJ et JOB BOOSTER) par région"]

# Effectif des bénéficiaires des projets financés (par le FAPE, le FASI, le FAIJ et JOB BOOSTER)par région selon le sexe
df_beneficiaires_projets = df[df['indicateur'] == "Effectif des bénéficiaires des projets financés (par le FAPE, le FASI, le FAIJ et JOB BOOSTER)par région selon le sexe"]

# Nombre d’emplois créés  par le FAPE, le FASI, le FAIJ, le PPEJ et JOB BOOSTER par région selon le sexe
df_emplois_crees = df[df['indicateur'] == "Nombre d’emplois créés  par le FAPE, le FASI, le FAIJ, le PPEJ et JOB BOOSTER par région selon le sexe"]
```

```python
# Répartition par sexe et par année 
import seaborn as sns
import matplotlib.pyplot as plt
df_grouped = df_demande_region_sexe.groupby(['Date', 'sexe'])['Value'].sum().unstack()
df_grouped.plot(kind='bar', figsize=(10, 6))
plt.title("Demandeurs d'emploi par année et sexe")
plt.ylabel("Nombre")
plt.xticks(rotation=45)
plt.legend(title="Sexe")
plt.tight_layout()
plt.show()
```

```python
# Nombre des demandeurs par tranche d’âge 
df_age_sum = df_demande_age.groupby('groupe-d-age')['Value'].sum().sort_values()
plt.figure(figsize=(10, 6))
df_age_sum.plot(kind='barh', color='skyblue')
plt.title("Demandeurs d'emploi par tranche d'âge")
plt.xlabel("Nombre")
plt.ylabel("Tranche d'âge")
plt.tight_layout()
plt.show()
```

```python
# Répartition des demandeurs d'emplois selon le diplome 
top_diplomes = df_demande_diplome.groupby('diplôme')['Value'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_diplomes.plot(kind='bar', color='orange')
plt.title("Top 10 des diplômes parmi les demandeurs d'emploi")
plt.ylabel("Nombre")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Nombre de postes d'emplois publiés par région
df_region_postes = df_postes_region.groupby('région')['Value'].sum().sort_values()
plt.figure(figsize=(10, 6))
df_region_postes.plot(kind='barh', color='green')
plt.title("Postes publiés par région")
plt.xlabel("Nombre de postes")
plt.tight_layout()
plt.show()
```

```python
# Répartition du nombre de financements sollicités et accordés par sexe 
df1 = df_financement_solicite.groupby('sexe')['Value'].sum()
df2 = df_financement_accorde.groupby('sexe')['Value'].sum()
df_compare = pd.DataFrame({"Sollicités (M FCFA)": df1,"Accordés (M FCFA)": df2})
df_compare.plot(kind='bar', figsize=(8, 6))
plt.title("Montants des financements sollicités vs accordés (par sexe)")
plt.ylabel("Montants (Millions FCFA)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Répartition des dix diplomes les plus demandés dans les offres d'emploi
df_diplome_postes = df_postes_diplome.groupby('diplôme')['Value'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
df_diplome_postes.plot(kind='bar', color='teal')
plt.title("Top 10 des diplômes demandés dans les offres d'emploi")
plt.ylabel("Nombre de postes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Nombre de projets financés par région
df_projets_region = df_projets_finances.groupby('région')['Value'].sum().sort_values()
plt.figure(figsize=(10, 6))
df_projets_region.plot(kind='barh', color='coral')
plt.title("Nombre de projets financés par région")
plt.xlabel("Nombre de projets")
plt.tight_layout()
plt.show()
```

```python
# Nombre de bénéficiaires des projets financés par région et par sexe (Année: 2022)
df_beneficiaires_2022 = df_beneficiaires_projets[df_beneficiaires_projets['Date'] == 2022]
plt.figure(figsize=(12, 6))
sns.barplot(data=df_beneficiaires_2022, x='région', y='Value', hue='sexe')
plt.title("Bénéficiaires des projets financés par région et par sexe (2022)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
